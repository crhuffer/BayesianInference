```python
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd
from scipy.stats import beta
from scipy import stats, optimize
```

$P(A|B) = \frac{P(B|A)P(A)}{P(B)}$

The likely hood function $L(p)$ is given by,

$L(p) \propto p^s(1-p)^f$ for $0<p<1$

Where $s$ and $f$ are the number of successes and failures respectively.


We update our belief, calculate our posterior probability, by multiplying our likelihood function by our prior belief:

$g(p|data) \propto g(p)L(p)$

```python
pvec = np.linspace(0.05, 0.95, 10); pvec
```

```python
priorvec = np.array([1, 5.2, 8, 7.2, 4.6, 2.1, 0.7, 0.1, 0, 0]); priorvec
```

```python
# normalize the priors

tempsum = priorvec.sum()
priorvec /= tempsum
priorvec
```

```python
fig, ax = plt.subplots()
plt.bar(x=pvec, height=priorvec, width = 0.05, color='b', label='Prior', alpha=0.5)
fig.legend()
ax.set_ylabel('Probability')
ax.set_xlabel('Outcome')
```

# Using the Beta distribution for your priors

https://en.wikipedia.org/wiki/Beta_distribution

The beta function is defined over the region [0, 1]

This makes it appropriate in problems like the ones we are discussing where you are mapping the probability of success.

Alpha and beta are different shape parameters of the beta function.

The term conjugate in this case means that the prior and the posterior follow the same distribution.

Beta is better for modeling than the discrete case that we used earlier because the actualy state space is continous and beta lets us handle that appropriately. There is no reason that the states should have to be at 0.05 or 0.15, but not inbetweeen these two values.

https://www.codeproject.com/Articles/56371/Finding-Probability-Distribution-Parameters-from-P
The above link describes how to determine beta and alpha for the beta function based on two percentiles.

Another interesting read:
https://en.wikipedia.org/wiki/Bayesian_inference


```python

def beta_parameters(x1, p1, x2, p2):
    "Find parameters for a beta random variable X so that P(X > x1) = p1 and P(X > x2) = p2."

    def square(x): 
        return x*x

    def objective(v):
        (a, b) = v
        temp  = square( stats.beta.cdf(x1, a, b) - p1 )
        temp += square( stats.beta.cdf(x2, a, b) - p2 )
        return temp

    # arbitrary initial guess of (3, 3) for parameters
    xopt = optimize.fmin(objective, (3, 3))
    return (xopt[0], xopt[1])   
```

```python
beta_parameters(.5, .9, .3, .5)
```

```python
Success = 11; Failure = 16

pvec = np.linspace(0,1,100)
Alpha, Beta = beta_parameters(.5, .9, .3, .5)
posterior = beta.pdf(pvec, Success+Alpha, Beta+Failure)
likelihood = beta.pdf(pvec, Success+1, Failure+1)
prior = beta.pdf(pvec, Alpha, Beta)

fig, ax = plt.subplots()

plt.plot(prior, '-r', label='prior')
plt.plot(posterior, '-b', label='posterior')
plt.plot(likelihood, '-k', label='likelihood')
plt.grid()
fig.legend()
ax.set_ylabel('Probability')
ax.set_xlabel('Outcome')
```
